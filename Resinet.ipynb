```python
import os
import numpy as np
class input_data:
    def __init__(self, img, label):
        self.img = img
        self.label = label
arr = []
path = 'tmp/with_mask/'
IMG_SIZE = 128
for img in os.listdir(path):
    path2 = path + img
    faces = RetinaFace.extract_faces(img_path = path2, align = False)
    for face in faces:
        img = np.array(face)
        face = cv2.resize(img,(IMG_SIZE, IMG_SIZE))
        data = input_data(face, 1)
        arr.append(data)
```

```python
import os
import numpy as np
class input_data:
    def __init__(self, img, label):
        self.img = img
        self.label = label
path = 'tmp/without_mask/'
IMG_SIZE = 128
for img in os.listdir(path):
    path2 = path + img
    faces = RetinaFace.extract_faces(img_path = path2, align = False)
    for face in faces:
        img = np.array(face)
        face = cv2.resize(img,(IMG_SIZE, IMG_SIZE))
        data = input_data(face, 0)
        arr.append(data)
```
